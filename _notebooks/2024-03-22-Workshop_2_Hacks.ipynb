## Question 1 - Pojos and Access Control:
Situation: The school librarian wants to create a program that stores all of the books within the library in a database and is used to manage their inventory of books available to the students. You decided to put your amazing code skills to work and help out the librarian!

a. Describe the key differences between the private and public access controllers and how it affects a POJO

Private access controllers restrict access to class members within the same class but public access controllers allow unrestricted access from any other class, facilitating interaction with the class's members. For a POJO, using private access for fields with public accessor and mutator methods ensures controlled access to the data.

b. Identify a scenario when you would use the private vs. public access controllers that isn't the one given in the scenario above

You would use private access controllers for the class's fields like name, age, and address but you would use public access controllers for methods like getName(), getAge(), and getAddress(), allowing other classes to retrieve information about a person.

c. Create a Book class that represents the following attributes about a book: title, author, date published, person holding the book and make sure that the objects are using a POJO, the proper getters and setters and are secure from any other modifications that the program makes later to the objects


```java
public class Book {
    private String title;
    private String author;
    private String datePublished;
    private String personHolding; 

    public String getTitle() {
        return title;
    }

    public void setTitle(String title) {
        this.title = title;
    }

    public String getAuthor() {
        return author;
    }

    public void setAuthor(String author) {
        this.author = author;
    }

    public String getDatePublished() {
        return datePublished;
    }

    public void setDatePublished(String datePublished) {
        this.datePublished = datePublished;
    }

    public String getPersonHolding() {
        return personHolding;
    }

    public void setPersonHolding(String personHolding) {
        this.personHolding = personHolding;
    }

    public static void printBooks(Book[] books) {
        for (Book book : books) {
            System.out.println("Title: " + book.getTitle() + ", Author: " + book.getAuthor());
        }
    }

    public static void main(String[] args) {
        Book[] books = new Book[3];
        books[0] = new Book();
        books[0].setTitle("Big Dawg");
        books[0].setAuthor("Me");

        books[1] = new Book();
        books[1].setTitle("Lil Bro");
        books[1].setAuthor("Paaras");

        books[2] = new Book();
        books[2].setTitle("Bro is the thinker");
        books[2].setAuthor("Varalu");

        printBooks(books);
    }
}
Book.main(null);
```

## Question 2 - Writing Classes:

(a) Describe the different features needed to create a class and what their purpose is.

Creating a basic class in Java involves these fundamental components:

1. **Class Declaration:**
   - Define the class using the `class` keyword followed by the class name.

2. **Fields (Instance Variables):**
   - Declare fields to represent the data or state of objects created from the class.
   - Fields define the attributes or properties associated with objects.

3. **Constructors:**
   - Create constructors to initialize objects of the class.
   - Constructors have the same name as the class and may accept parameters to initialize object state.

4. **Methods:**
   - Methods are for the logic associated with the class, such as calculations or interactions with other objects.

(b) Code:

Create a Java class BankAccount to represent a simple bank account. This class should have the following attributes:
- accountHolder (String): The name of the account holder.
balance (double): The current balance in the account.
Implement the following mutator (setter) methods for the BankAccount class:
- setAccountHolder(String name): Sets the name of the account holder.
- deposit(double amount): Deposits a given amount into the account.
- withdraw(double amount): Withdraws a given amount from the account, but only if the withdrawal amount is less than or equal to the current balance.
Ensure that the balance is never negative.

```java
public class BankAccount {
    private String accountHolder;
    private double balance;

    public BankAccount(String accountHolder, double balance) {
        this.accountHolder = accountHolder;
        this.balance = balance;
    }

    public void setAccountHolder(String accountHolder) {
        this.accountHolder = accountHolder;
    }

    public double getBalance() {
        return balance;
    }

    // Method to deposit money into the account
    public void deposit(double amount) {
        if (amount > 0) {
            balance += amount;
            System.out.println(amount + " deposited successfully.");
        } else {
            System.out.println("Invalid deposit amount.");
        }
    }

    // Method to withdraw money from the account
    public void withdraw(double amount) {
        if (amount > 0 && amount <= balance) {
            balance -= amount;
            System.out.println(amount + " withdrawn successfully.");
        } else {
            System.out.println("Withdrawal amount exceeds balance or is invalid.");
        }
    }

    public static void main(String[] args) {
        BankAccount account = new BankAccount("John Doe", 1000.0);
        
        System.out.println("Initial Balance: " + account.getBalance());
        
        account.deposit(500.0);
        System.out.println("Current Balance after deposit: " + account.getBalance());
        
        account.withdraw(200.0);
        System.out.println("Current Balance after withdrawal: " + account.getBalance());
    }
}
BankAccount.main(null);
```

## Question 3 - Instantiation of a Class

(a) Explain how a constructor works, including when it runs and what generally is done within a constructor.

A constructor is as a special method responsible for initializing objects of a class. It works automatically when you create an object using the `new` keyword. When an object is instantiated, the constructor with the same name as the class is invoked, ensuring that the object starts with a defined initial state. Constructors facilitate the process of setting initial values to the object's fields, ensuring that the object is in a consistent and usable state right from its creation.

(b) Create an example of an overloaded constructor within a class. You must use at least three variables. Include the correct initialization of variables and correct headers for the constructor. Then, run the constructor at least twice with different variables and demonstrate that these two objects called different constructors. 

```java
public class Person {
    private String name;
    private int age;
    private String occupation;

    public Person() {
        this.name = "Unknown";
        this.age = 0;
        this.occupation = "Unemployed";
    }

    public Person(String name, int age, String occupation) {
        this.name = name;
        this.age = age;
        this.occupation = occupation;
    }

    public void displayInfo() {
        System.out.println("Name: " + name + ", Age: " + age + ", Occupation: " + occupation);
    }

    public static void main(String[] args) {
        Person person1 = new Person();
        Person person2 = new Person("Alice", 30, "Engineer");

        System.out.println("Person 1:");
        person1.displayInfo();

        System.out.println("\nPerson 2:");
        person2.displayInfo();
    }
}
Person.main(null);
```

## Question 4 - Wrapper Classes:

(a) Provide a brief summary of what a wrapper class is and provide a small code block showing a basic example of a wrapper class.

A wrapper class in Java is a class that encapsulates, or wraps, primitive data types to provide them with functionalities of objects. It allows primitive types to be used as objects and provides additional methods and utilities.

(b) Create a Java wrapper class called Temperature to represent temperatures in Celsius. Your Temperature class should have the following features:

Fields:

A private double field to store the temperature value in Celsius.


Constructor:

A constructor that takes a double value representing the temperature in Celsius and initializes the field.


Methods:

getTemperature(): A method that returns the temperature value in Celsius.
setTemperature(double value): A method that sets a new temperature value in Celsius.
toFahrenheit(): A method that converts the temperature from Celsius to Fahrenheit and returns the result as a double value. 

```java
public class Temperature {
    private double temperatureCelsius;

    public Temperature(double temperatureCelsius) {
        this.temperatureCelsius = temperatureCelsius;
    }

    public double getTemperature() {
        return temperatureCelsius;
    }

    public void setTemperature(double value) {
        this.temperatureCelsius = value;
    }

    public double toFahrenheit() {
        return (temperatureCelsius * 9 / 5) + 32;
    }

    public static void main(String[] args) {
        Temperature temp = new Temperature(20.0);

        System.out.println("Temperature in Celsius: " + temp.getTemperature());
        System.out.println("Temperature in Fahrenheit: " + temp.toFahrenheit());

        temp.setTemperature(25.0);

        System.out.println("\nUpdated Temperature in Celsius: " + temp.getTemperature());
        System.out.println("Updated Temperature in Fahrenheit: " + temp.toFahrenheit());
    }
}
Temperature.main(null);
```

## Question 5 - Inheritence:

Situation: You are developing a program to manage a zoo, where various types of animals are kept in different enclosures. To streamline your code, you decide to use inheritance to model the relationships between different types of animals and their behaviors.

(a) Explain the concept of inheritance in Java. Provide an example scenario where inheritance is useful.

Inheritance allows a subclass to inherit attributes and methods from a superclass, so that you can reuse your code easily. For example, a superclass "Animal" could have attributes like "name" and "age," and methods like "eat()" and "sleep()," while subclasses like "Dog" and "Cat" can inherit these traits and behaviors while adding their own unique functionalities, such as "bark()" for dogs and "purr()" for cats.

(b) Code:

You need to implement a Java class hierarchy to represent different types of animals in the zoo. Create a superclass Animal with basic attributes and methods common to all animals, and at least three subclasses representing specific types of animals with additional attributes and methods. Include comments to explain your code, specifically how inheritance is used.

```java
public class Animal {
    private String name;
    private int age;
    private String habitat;

    public Animal(String name, int age, String habitat) {
        this.name = name;
        this.age = age;
        this.habitat = habitat;
    }

    public String getName() {
        return name;
    }

    public int getAge() {
        return age;
    }

    public String getHabitat() {
        return habitat;
    }

    public void eat() {
        System.out.println(name + " is eating.");
    }

    public void sleep() {
        System.out.println(name + " is sleeping.");
    }
}

public class Liger extends Animal {
    private String parent1;
    private String parent2;

    public Liger(String name, int age, String habitat, String parent1, String parent2) {
        super(name, age, habitat); // Call to superclass constructor
        this.parent1 = parent1;
        this.parent2 = parent2;
    }

    public void roar() {
        System.out.println(getName() + " the Liger is roaring!");
    }
}

public class Elephant extends Animal {
    private String herdName;

    public Elephant(String name, int age, String habitat, String herdName) {
        super(name, age, habitat); // Call to superclass constructor
        this.herdName = herdName;
    }

    public void trumpet() {
        System.out.println(getName() + " the Elephant is trumpeting!");
    }
}

public class Meerkat extends Animal {
    private int numFamilyMembers;

    public Meerkat(String name, int age, String habitat, int numFamilyMembers) {
        super(name, age, habitat); // Call to superclass constructor
        this.numFamilyMembers = numFamilyMembers;
    }

    public void dig() {
        System.out.println(getName() + " the Meerkat is digging!");
    }
}
```
