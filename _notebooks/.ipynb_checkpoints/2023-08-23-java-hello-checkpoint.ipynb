```java
// Define a class for a Person
class Person {
    // Instance variables (attributes)
    private String name;
    private int age;

    // Constructor to initialize name and age
    public Person(String name, int age) {
        this.name = name;
        this.age = age;
    }

    // Getter methods to retrieve name and age
    public String getName() {
        return name;
    }

    public int getAge() {
        return age;
    }

    // Setter methods to update name and age
    public void setName(String name) {
        this.name = name;
    }

    public void setAge(int age) {
        this.age = age;
    }
}

// Define a class for Dessert
class Dessert {
    // Instance variables (attributes)
    private String type;
    private double cost;

    // Constructor to initialize type and cost
    public Dessert(String type, double cost) {
        this.type = type;
        this.cost = cost;
    }

    // Getter methods to retrieve type and cost
    public String getType() {
        return type;
    }

    public double getCost() {
        return cost;
    }

    // Setter methods to update type and cost
    public void setType(String type) {
        this.type = type;
    }

    public void setCost(double cost) {
        this.cost = cost;
    }
}

// Define a class for Location
class Location {
    // Instance variables (attributes)
    private String city;
    private String state;
    private String zip;

    // Constructor to initialize city, state, and zip
    public Location(String city, String state, String zip) {
        this.city = city;
        this.state = state;
        this.zip = zip;
    }

    // Getter methods to retrieve city, state, and zip
    public String getCity() {
        return city;
    }

    public String getState() {
        return state;
    }

    public String getZip() {
        return zip;
    }

    // Setter methods to update city, state, and zip
    public void setCity(String city) {
        this.city = city;
    }

    public void setState(String state) {
        this.state = state;
    }

    public void setZip(String zip) {
        this.zip = zip;
    }
}

public class Main {
    public static void main(String[] args) {
        // Create instances of the Person class
        Person person1 = new Person("Alice", 30);
        Person person2 = new Person("Bob", 25);

        // Display initial information about persons
        System.out.println("Person 1: Name - " + person1.getName() + ", Age - " + person1.getAge());
        System.out.println("Person 2: Name - " + person2.getName() + ", Age - " + person2.getAge());

        // Use setter methods to change information about persons
        person1.setName("Alicia");
        person2.setAge(26);

        // Display updated information about persons
        System.out.println("\nUpdated Person 1: Name - " + person1.getName() + ", Age - " + person1.getAge());
        System.out.println("Updated Person 2: Name - " + person2.getName() + ", Age - " + person2.getAge());

        // Create instances of the Dessert class
        Dessert dessert1 = new Dessert("Cake", 10.99);
        Dessert dessert2 = new Dessert("Ice Cream", 4.50);

        // Display information about desserts
        System.out.println("\nDessert 1: Type - " + dessert1.getType() + ", Cost - $" + dessert1.getCost());
        System.out.println("Dessert 2: Type - " + dessert2.getType() + ", Cost - $" + dessert2.getCost());

        // Create instances of the Location class
        Location location1 = new Location("New York", "NY", "10001");
        Location location2 = new Location("Los Angeles", "CA", "90001");

        // Display information about locations
        System.out.println("\nLocation 1: City - " + location1.getCity() + ", State - " + location1.getState() + ", Zip - " + location1.getZip());
        System.out.println("Location 2: City - " + location2.getCity() + ", State - " + location2.getState() + ", Zip - " + location2.getZip());
    }
}
```
